```python
import pandas as pd
```

```python
mydata=pd.read_csv(r"C:\Users\dell\Desktop\ML\DataSets-master\BMI_male_female.csv")
```

```python
mydata.head()
```

```python
mydata["Gender"]=mydata["Gender"].map({"Male":0,"Female":1})
```

```python
mydata.head(1)
```

```python
mydata.describe()
```

```python
x_feature=mydata.drop("Index",axis=1)
y_target=mydata["Index"]
```

```python
Index_Name=pd.Series(["Extremely Weak","Weak","Normal","Over Weight","Obesity","Extreme Obesity"])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_feature,y_target,test_size=.3,random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
trainer=DecisionTreeClassifier(criterion="entropy")
```

```python
learner=trainer.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
#xtest....y_test
ya=y_test
yp=learner.predict(x_test)
acc=accuracy_score(ya,yp)*100
acc
```
